```python
import pandas as pd
import numpy as np
```

```python
# LOAD AND COMBINE 
import glob
path = 'C:\\Users\\Nyla Majid\\OneDrive\\Documents\\cyclist project'
all_files = glob.glob(path + "/*.csv")

data= pd.concat((pd.read_csv(f) for f in all_files), ignore_index=True)
data.head()
```

```python
data.shape
```

```python
#DROP ROWS WITH MISSING INFO
data.dropna(subset=['start_station_name','end_station_name','started_at','ended_at'],inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
#CONVERTING DATA TYPES
data['started_at']=pd.to_datetime(data['started_at'])
data['ended_at']=pd.to_datetime(data['ended_at'])
data.dtypes
```

```python
#RIDE LENGTH IN MINUTES
data['ride_length']=(data['ended_at'] - data['started_at']).dt.total_seconds()/60
data.head(3)
```

```python
#filtering invalid ride
data=data[data['ride_length']>0]
```

```python
# ADDINH TIME FEATURES
data.loc[:,'weekday']=data['started_at'].dt.day_name()
data.loc[:,'month']=data['started_at'].dt.month_name()
data.loc[:,'hour']=data['started_at'].dt.hour
data.head(3)
```

```python
#SAVING CLEAN DATA SET FOR POWERBI
data.to_csv('C:\\Users\\Nyla Majid\\OneDrive\\Documents\\cyclist project\\cyclist_cleaned.csv',index = False)
```
